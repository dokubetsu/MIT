```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random
```

```python
df = pd.read_csv('Book2.csv')
df
```

```python
df1=df[['x']]
df1
```

```python
df2=df[['y']]
df2
```

```python
X=df1.to_numpy()
X
```

```python
Y=df2.to_numpy()
Y
```

```python
model = LinearRegression()
model
```

```python
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X)
```

```python
model.predict([[7]])
```

```python
model.predict([[12]])
```

```python
plt.title("Scatter diagram")
plt.scatter(X,Y)
for i, j in zip(X, Y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

```python
plt.scatter(X, Y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
for i, j in zip(X, Y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

# Using    Random

```python
## Integers
```

```python
random1=[]
for i in range(0,5):
    random1.append( random.randint(1,30))
x1=np.array(random1)
x1
```

```python
random2=[]
for i in range(0,5):
    random2.append( random.randint(1,30))
y1=np.array(random2)
y1
```

```python
model1 = LinearRegression()
model1
```

```python
model1.fit(x1.reshape(-1,1),y1)
```

```python
model1.coef_
```

```python
model1.intercept_
```

```python
y_pred1 = model1.predict((x1.reshape(-1,1)))
```

```python
plt.title("Scatter diagram")
plt.scatter(x1,y1)
for i, j in zip(x1, y1):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

```python
plt.scatter(x1, y1, color='blue')
plt.plot(x1, y_pred1, color='red')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('Linear Regression')
for i, j in zip(x1, y1):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```
