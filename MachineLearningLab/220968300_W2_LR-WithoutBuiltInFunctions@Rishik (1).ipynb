```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('Book2.csv')
df
```

```python
df1=df[['x']]
df1
```

```python
df2=df[['y']]
df2
```

```python
X=df1.to_numpy()
X
```

```python
Y=df2.to_numpy()
```

```python
plt.title("Scatter diagram")
plt.scatter(X,Y)
for i, j in zip(X, Y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

```python
xmean=X.mean()
xmean
```

```python
ymean=Y.mean()
ymean
```

```python
p=np.multiply(X,Y)
pmean=p.mean()
pmean
```

```python
ssq=X*X
ssqmean=ssq.mean()
ssqmean
```

```python
a1=((pmean)-(xmean*ymean))/(ssqmean-(xmean*xmean))
a1
```

```python
a0=ymean-(a1*xmean)
a0
```

```python
y_pred = a1 * X + a0
```

```python
plt.title("Plot")
plt.scatter(X,Y)
plt.scatter(X, Y, color = 'red')
plt.plot(X,y_pred,color="blue")
for i, j in zip(X, Y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

```python
#7th Week Sales(X=7)
y_pred=a1*7+a0
y_pred
```

```python
#12th Week Sales(X=12)
y_pred=a1*12+a0
y_pred
```
