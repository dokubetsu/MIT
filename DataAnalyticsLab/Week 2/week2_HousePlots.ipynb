```python
import pandas as pd
import matplotlib.pyplot as plt

house_prices_df = pd.read_csv('Bengaluru_House_Prices.csv')

#1. Compute the total number of missing values in the DataFrame.

total_missing = house_prices_df.isnull().sum().sum()
print('Total values missing: ',total_missing)
print('All values: ',len(house_prices_df))

#Compute the percentage of missing values in the DataFrame.

percentage_missing=(total_missing/(len(house_prices_df)))*100
print('Percentage of missing values is: ',percentage_missing)
```

```python
#If a column has more than 15% missing values then drop (or remove) the column from the DataFrame.
columns_to_drop = house_prices_df.columns[house_prices_df.isnull().sum() / len(house_prices_df) > 0.15]
house_prices_df.drop(columns_to_drop, axis=1, inplace=True)
house_prices_df.head(10)
```

```python
#3. Which column has the most number of missing values?
column_with_most_missing = house_prices_df.isnull().sum().idxmax()
print("Column with the most missing values:", column_with_most_missing)
```

```python
#4. drop (or remove) the rows containing the missing values in the location, size, and total_sqft columns
house_prices_df.dropna(subset=['location', 'size', 'total_sqft'], inplace=True)
house_prices_df.head(10)
```

```python
#5. drop all the rows in the bath column containing more than 5 bathrooms
house_prices_df = house_prices_df[house_prices_df['bath'] <= 5]
house_prices_df.head(10)
```

```python
#6. Find the houses available in each area.
houses_available_per_area = house_prices_df['location'].value_counts()
print(houses_available_per_area)
```

```python
#7. Find the top five areas where the large number of houses available
top_five_areas = houses_available_per_area.head(5)
print("Top five areas with the largest number of houses available:\n", top_five_areas)
```

```python
#8. Visualize the house price based on build up area of top five locations where the large number of houses available (Visualize by independent charts.)
top_five_areas = houses_available_per_area.head(5)
top_five_locations = top_five_areas.index.tolist()
for location in top_five_locations:
    location_df = house_prices_df[house_prices_df['location'] == location]
    plt.scatter(location_df['total_sqft'], location_df['price'])
    plt.title(f"House Price vs Build-up Area in {location}")
    plt.xlabel("Total Sqft")
    plt.ylabel("Price")
 
    plt.show()
```
