```python
import pandas as pd
import numpy as np
df=pd.read_excel('Cereals1.xls')
```

```python
#1.Create a table with the 5-number summary of all the numeric attributes
df.describe()
```

```python
#2.For each of the numeric attributes (proteins up to vitamins), identify and replace all 
    #missing data (indicated with -1) with the arithmetic mean of the attribute.


numeric_attributes=['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','weight','cups','rating']
df=df.mask(df==-1).fillna(df.mean())
df.describe()
```

```python
#3.

#Yes, the strategy used in dealing with missing values was effective
```

```python
#Calulating the inter quartile range of each numeric attibute.
for attribute in numeric_attributes:
    q1 = df[attribute].quantile(0.25)
    q3 = df[attribute].quantile(0.75)
    iqr = q3 - q1
    print(attribute,iqr)
```

```python
#Calculating the lower limit for each numeric attribute
for attribute in numeric_attributes:
    q1 = df[attribute].quantile(0.25)
    q3 = df[attribute].quantile(0.75)
    iqr = q3 - q1
    lower_limit=q1-1.5*iqr
    print(attribute,lower_limit)
```

```python
#Calculating the upper limit for each numeric attribute
for attribute in numeric_attributes:
    q1 = df[attribute].quantile(0.25)
    q3 = df[attribute].quantile(0.75)
    iqr = q3 - q1
    upper_limit=q3+1.5*iqr
    print(attribute,upper_limit)
```

```python
#4.For each of the numeric attributes, identify and replace all noisy data with the median of attribute. 
for attribute in numeric_attributes:
    q1 = df[attribute].quantile(0.25)
    q3 = df[attribute].quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr
    mask = (df[attribute] < lower_limit) | (df[attribute] > upper_limit)
    median_value = df[attribute].median()
    df.loc[mask, attribute] = median_value
df.describe()
```

```python
#The strategy used in dealing with noisy values was effective.
```
