```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('mtcars.csv')
```

```python
#1. Find the car with the best mpg

best_mpg = df[df['mpg'] == df['mpg'].max()]
print('Car with best milage: ',best_mpg)
```

```python
#2. Find the car with the worst mpg

worst_mpg = df[df['mpg'] == df['mpg'].min()]
print('nCar with worst milage: ',worst_mpg)
```

```python
#3. Find the car with the best horsepower

best_hp = df[df['hp'] == df['hp'].max()]
print('Car with best horsepower: ',best_hp)
```

```python
#4. Find 5 number summary of displacement

displacement_summary = df['disp'].describe()
print('5 number summmary table for Displacement: ',displacement_summary)
```

```python
#5. Find median horse power

median_hp = df['hp'].median()
print(' Median horsepower: ',median_hp)
```

```python
#6. What is average mpg for manual vs. automatic cars

avg_mpg_by_transmission = df.groupby(by='am').mpg.mean()
print('Mean mpg of the two transmission types: ',avg_mpg_by_transmission)
```

```python
#7. Draw a histogram of miles per gallon

sns.histplot(df['mpg'])
plt.title('Histogram of Miles Per Gallon')
plt.xlabel('Miles Per Gallon')
plt.ylabel('Frequency')
plt.show()
```

```python
#8. Boxplot of mpg for each cylinder type

sns.boxplot(x='cyl', y='mpg', data=df)
plt.title('Boxplot of MPG by Cylinder Type')
plt.xlabel('Cylinders')
plt.ylabel('Miles Per Gallon')
plt.show()
```

```python
#9. Create a crosstab displaying count of automatic vs. manual cars

crosstab_am = pd.crosstab(df.am,df.am)
print('Crosstab of Transmission type: ',crosstab_am)
```

```python
#10. Create a crosstab displaying count of “am vs cyl”

crosstab_am_cyl = pd.crosstab(index=df['am'], columns=df['cyl'])
print('Crosstab transmission type vs number of cylinders: ',crosstab_am_cyl)
```

```python
#11. What is the correlation between the weight of the car and mpg

correlation_weight_mpg = df['wt'].corr(df['mpg'])
print('Correlation between weight and mpg: ',correlation_weight_mpg)
```
