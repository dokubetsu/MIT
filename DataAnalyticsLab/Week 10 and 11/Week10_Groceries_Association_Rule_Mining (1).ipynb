```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_table('groceries.csv',header=None)
df
```

```python
#2. Prepare the data for finding association rules. Each transaction will contain a list of item #in the transaction.
#[['citrus fruit', 'semi-finished bread', 'margarine', 'ready soups'],['tropical fruit', 'yogurt', 'coffee'],…….['whole milk']]
```

```python
df[0] = df[0].apply(lambda x : x.split(","))
df
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#3. Use Python library mlxtend and convert the transactions into a format that can be used 
#   in the Apriori method for finding frequent itemsets. 

te = TransactionEncoder()
te_ary = te.fit(df[0]).transform(df[0])
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
#4. Find top selling items with minimum support of 2%.

apriori(df, min_support=0.02,use_colnames = True).sort_values(by = "support" , ascending = False).head()
```

```python
#5. Find all frequent itemsets with minimum support of 5%

apriori(df, min_support=0.05,use_colnames = True).sort_values(by = "support")
```

```python
#6. Find all frequent itemsets of length 2 with minimum support of 2%

frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[frequent_itemsets['length']==2]
```

```python
#7. Find the top 10 association rules with minimum support of 2%, sorted by confidence in descending order.

var = apriori(df, min_support=0.02, use_colnames=True)
var = association_rules(var, metric="lift")
var.sort_values(by="confidence" , ascending =False)
```

```python
#8. Find association rules with minimum support of 2% and lift of more than 1.0.

var[var['lift']>1]
```
